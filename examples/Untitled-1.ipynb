```python
from examples import construct_sphere as c
from utils.sphere import Sphere
import pyshtools as pysh

from utils.plotting import plot_polar_rad
```

```python
import numpy as np

sphere = Sphere()
np.unique(sphere.latitudes)
```

```python
lat_coeffs = sphere.sh_coeffs_lats()
lon_coeffs = sphere.sh_coeffs_lons()
r_coeffs = sphere.sh_coeffs_radius()
```

```python
print(r_coeffs.shape)
print(r_coeffs)
```

```python
# Reconstruct sphere without noise

lat_recon = pysh.expand.MakeGridPoint(lat_coeffs, lat=sphere.latitudes, lon=sphere.longitudes)
lon_recon = pysh.expand.MakeGridPoint(lon_coeffs, sphere.latitudes, sphere.longitudes)
r_recon = pysh.expand.MakeGridPoint(r_coeffs, sphere.latitudes, sphere.longitudes)

plot_polar_rad(r_recon, lon_recon, lat_recon)
```

```python
# Reconstruct with noise

c_lat_noisy = c.add_noise_to_coefficients(lat_coeffs, 0.001)
c_lon_noisy = c.add_noise_to_coefficients(lon_coeffs, 0.001)
c_r_noisy = c.add_noise_to_coefficients(r_coeffs, 0.1)

lat_noisy = pysh.expand.MakeGridPoint(c_lat_noisy, lat=sphere.latitudes, lon=sphere.longitudes)
lon_noisy = pysh.expand.MakeGridPoint(c_lon_noisy, sphere.latitudes, sphere.longitudes)
r_noisy = pysh.expand.MakeGridPoint(c_r_noisy, sphere.latitudes, sphere.longitudes)

plot_polar_rad(r_noisy, lat_noisy, lon_noisy)

```
